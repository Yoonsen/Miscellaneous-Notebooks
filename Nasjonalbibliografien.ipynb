```python
korpus_subject = mm.metadata(urns=korpus_urn, marctuples = [(653, ' ', ' ', 'a'), (655,' ', 7, 'a')] )
```

```python
subjects = nb.frame(korpus_subject)
subjects
```

```python
nullindex = subjects[subjects['653- - -a'].isnull() & subjects['655- -7-a'].isnull()].index
```

```python
non_null = [x for x in subjects.index if not x in nullindex]
```

```python
subjects.loc[non_null]
```

```python
nøkkelord = list(subjects.loc[non_null]['653- - -a'].dropna().values)
nøkkelord
```

```python
opptelling = Counter()
nøkkelordslister = [klassifikasjon.split(', ') for klassifikasjon in nøkkelord]
for x in nøkkelordslister:
    opptelling.update(x)
```

```python
nb.frame_sort(nb.frame(opptelling)).head(50)
```

```python
nøkkelord2 = list(subjects.loc[non_null]['655- -7-a'].dropna().values)
nøkkelord2
```

```python
opptelling2 = Counter()
nøkkelordslister2 = [klassifikasjon.split(', ') for klassifikasjon in nøkkelord2]
for x in nøkkelordslister2:
    opptelling2.update(x)
```

```python
nb.frame_sort(nb.frame(opptelling2)).head(50)
```
