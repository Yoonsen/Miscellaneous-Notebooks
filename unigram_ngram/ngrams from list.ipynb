```python
import dhlab.nbtext as nb
import matplotlib.pyplot as plt
import pandas as pd
#plt.rcParams.update({'font.size': 24})
```

```python
def ngrams(grams, years =(1800, 2020), corpus = 'bok', column = None):
    res = pd.concat([nb.nb_ngram(n, years = years, corpus = corpus) for n in grams], axis = 1)
    #if column != None:
    #    res.columns = [column]
    #else:
    #    res.columns = [grams[0]]
    return res
```

```python
def truncated(word, factor = 10, freq = 50, limit = 50):
    return list(nb.sorted_wildcardsearch(
    {
        'word': word,   # her legges selve søkeordet inn
        'factor':factor,           # factor bestemmer hvor mye lenger treffene skal være enn ordet med jokertegn.
        'freq_lim':freq,        # sett begrensninger på frekvensen, minimumsverdi
        'limit':limit           # begrensning på antall treff
        
    }).index)
```

```python
tele = list(set([x.lower() for x in truncated('Telegraf*') + truncated('telegraf*')]))
```

```python
a = ngrams(tele)
```

```python
a.loc[1865:1975].plot(legend=False)
```

```python
a[['telegrafiske', 'telegrafstationen', 'telegraf', 'telegrafiske']].plot()
```
