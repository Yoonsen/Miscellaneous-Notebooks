```python
import dhlab.nbtext as nb
import pandas as pd
```

```python
tot = nb.frame(nb.totals(100000), 'tot')
nb.normalize_corpus_dataframe(tot)
```

```python
nb.book_urn(title = 'jonas')
```

```python
def coll(word, comp = tot, urn = 2011071805547):
    coll = nb.urn_coll(word, urns = [urn], before = 5, after = 5)
    coll.columns = [word]
    nb.normalize_corpus_dataframe(coll)
    return nb.frame_sort(pd.concat([coll, nb.frame(coll[word]/tot['tot'], 'coll')], axis = 1), 'coll')
```
