```python
def uni(words,  **kwargs):
    r = nb.unigram('.', **kwargs)
    if isinstance(words, str):
        ws = nb.frame(nb.unigram(words, **kwargs)/r, words)
    else:
        # list or keys or something
        ws = pd.concat([nb.frame(nb.unigram(word, **kwargs)/r, word) for word in words], axis=1, sort=True)
            
    return ws
```
