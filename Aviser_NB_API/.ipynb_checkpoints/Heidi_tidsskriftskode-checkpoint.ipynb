```python
import dhlab.module_update as mu
mu.update('conc_coll_mediatype')
import conc_coll_mediatype as cc
import dhlab.nbtext as nb
import pandas as pd

import requests

from bs4 import BeautifulSoup

pd.set_option('max_colwidth', 1000)
```

```python
def metadata(urn, key = 'title'):
    m = meta_urn(urn)
    soup = BeautifulSoup(m)
    title = ""
    for x in soup.find_all(key):
        title += " " + x.text
    return title
```

```python
def meta_urn(urn):
    res = requests.get("https://api.nb.no/catalog/v1/metadata/{id}/mods".format(id = urn))
    return res.text
    
```

```python
def vis_konk(k):
    df = nb.frame(k)
    df['url'] = "<a target = '_' href='https://nb.no/items/" + df['urn'].str[:-5] + "'>klikk!</a>"
    return df[['url', 'before', 'word', 'after']].style
```
