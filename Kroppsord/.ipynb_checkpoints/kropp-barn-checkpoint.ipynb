```python
def uni(words,  **kwargs):
    r = nb.unigram('.', **kwargs)
    if isinstance(words, str):
        ws = nb.frame(nb.unigram(words, **kwargs)/r, words)
    else:
        # list or keys or something
        ws = pd.concat([nb.frame(nb.unigram(word, **kwargs)/r, word) for word in words], axis=1, sort=True)
            
    return ws
```

```python
kroppsord_sing = "hode panne øye øyebryn øyevipper kinn hake nese leppe øre nakke hals skulder bryst mave mage hofte lår legg fot tå arm underarm hånd tommel finger langfinger ringfinger".split()
```

```python
forms = lambda x:[x for x in nb.word_paradigm(x) if x[0] == 'subs'][0][1]
```

```python
nb.word_variant('arm','be')
```

```python
def bestemt_ent(w):
    #forms = [(x[0].split(), x[1]) for x in nb.word_variant(w,'be')]
    res = [x for x in nb.word_variant(w,'be') if 'subst' in x[0] and 'ent' in x[0]]
    return res
```

```python
def bestemt_flertall(w):
    #forms = [(x[0].split(), x[1]) for x in nb.word_variant(w,'be')]
    res = [x for x in nb.word_variant(w,'be') if 'subst' in x[0] and 'fl' in x[0]]
    return res
```

```python
bestemt_flertall('bil')
```

```python
kord = {x:forms(x) for x in kroppsord_sing}
```

```python
kroppsord_best ={f[1] for k in kroppsord_sing for f in bestemt_ent(k)}
```

```python
kroppsord_best
```

```python
def frameit(p, x):
    try:
        r = nb.frame(p, x)
    except:
        r = p
    return r
```

```python
res = pd.concat([frameit(nb.bigram(first = x, second='hennes', period = (1910, 2010)), x) for x in kroppsord_best], axis = 1)
```
