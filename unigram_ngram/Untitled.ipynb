```python
import dhlab.nbtext as nb
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 24})
```

```python
nb.nb_ngram("Kjærlighed", years = (1800, 1950)).plot(lw=5, figsize=(15,8))
```

```python
nb.nb_ngram("Kjærlighed, Kjcerlighed", years = (1800, 1950)).sort_index().rolling(window=5).mean().plot(lw=5, figsize=(15,8))
```

```python
nb.nb_ngram("Kjærlighed, Kjcerlighed, Kjærlighet", years = (1800, 1950)).sort_index().plot(lw=5, figsize=(15,8))
```

```python
p = nb.nb_ngram("Kjærlighed, Kjcerlighed, Kjoerlighed", years = (1800, 1950)).sort_index().sum(axis=1)
p = nb.frame(p, 'Kjærlighed')
p.plot(lw=5, figsize=(15,8))
```
