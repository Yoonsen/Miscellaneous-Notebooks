# Konkordanser (nøkkelord i kontekst): liste over forekomster av ord, inkludert kontekst, i et digitalt korpus

# For å sjekke konkordanser i bøker og aviser benyttes kommandoen concordance()

## 1. Periodeavgrensede konkordanser fra Bokhyllas aviskorpus

```python
nb.concordance(word=['hvalfangst'], corpus='avis', title = '%', yearfrom=1880, yearto = 1930, 
               before = 12, after = 12)
```

```python
nb.concordance(word=['pancreas'], corpus='avis', title = '%', yearfrom=1880, yearto = 1930, 
               before = 12, after = 12)
```
