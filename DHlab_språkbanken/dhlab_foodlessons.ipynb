## Korpus 

En samling referanser til tekster. Samme som nettbiblioteket


```python
word1 = 'mat'
word2 = 'drikke'
startår = 1960
sluttår = 2010

corpus1 = d2.document_corpus(subject = word1, doctype = 'digibok', from_year = startår, to_year = sluttår, limit = 50, lang = 'nob')
corpus2 = d2.document_corpus(subject = word2, doctype = 'digibok', from_year = startår, to_year = sluttår, limit = 50, lang = 'nob')

c1_agg = d2.get_document_frequencies(list(corpus1.urn))
c2_agg = d2.get_document_frequencies(list(corpus2.urn))

combo = pd.concat([c1_agg.sum(axis = 1), c2_agg.sum(axis = 1)], axis = 1)
combo.columns = [word1, word2]
```

```python
corpus1.sample(5)["urn title authors city year publisher subjects ddc literaryform".split()]
```
