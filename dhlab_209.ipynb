# Dhlab 

Dhlab har en ny struktur og en litt annen organisering av kode, men ikke så mye. Det er nye oppdaterte databaser, aviser, bøker og annet.

Fremfor å ha funksjoner, er det må mer fokus på at brukeren bygger objekter som kan vises og inngå som byggesteinger i andre objekter. For eksempel et korpus inngår i konkordanser og temamodeller med mer.

Toppnivået er `dhlab` som før, og alt av tidligere notebøker skal virke. Men fint å gå over til de nyere versjonene som vist under.

## Korpus og konkordans

La oss starte med å bygge et korpus for så å søke i det. Det kan komme fra mange steder. Her kan vi først prøve å bygge med metadat internt i dhlab.

````python
import dhlab.text as dh
````

Med `dh` kan du nå korpusbygging og data fra korpus

````python
sandel = dh.Corpus(author = "Cora Sandel", limit = 20)
````

Bruk shift+tab for å se alt som kan inngå av parametre. Her er en liste
```
dh.Corpus(
    doctype=None, for eksempel "digibok", "digavis", "digitidsskrift", "digistorting"
    author=None,  Skriv noen ord, ikke følsomt for kapitalisering
    freetext=None, At bøkene skal inneholde bestemte ting - eksperimenter!
    from_year=None,
    to_year=None,
    from_timestamp=None, (timestamp for aviser 20020105 (5 januar 2002)
    to_timestamp=None,
    title=None, innholdsord i tittel
    ddk=None,   dewey desimalnummer  - mulig det ikke fungerer så godt
    subject=None, temaord fra biblioteket
    lang=None, trebokstavode eng engelsk, nob bokmål, nno nynorsk sme nordsamisk ogs.
    limit=None, antall bøker

)
```

Korpuset har et attributt `corpus` som er en dataramme. Den viser innholdet og en del ekstra info

````python
sandel.corpus
````

Fra korpuset kan du hente ut konkordanser og kollokasjoner, første argument er et korpus (som over) eller en pandas dataramme som inneholder en kolonne med navn urn. 

Med den nye indeksen er det kommet nye måter å søke etter passasjer. Her har jeg gitt et forholdsvis komplekst eksempel med nærhetssøk. Alle søkene går innenfor et avsnitt, uansett hvor stor konteksten angis.

````python
konk = dh.Concordance(sandel, "NEAR(alberte jacob, 5) OR NEAR(alberte frykt*, 10)")
konk.size
````

Avhening av størrelsen på konkordansen, kan `konk.show()` kjøres flere ganger. Antallet konkordanser angis som argument til `show()`. Her er det bare å eksperimentere! 

Bygg gjerne forskjellige konkordanseobjekter for sammenligning. 

````python
konk.show(5)
````

Konkordansene kan også lagres til excel

````python
konk.concordance.to_excel('sandel.xlsx')
````

Kollokasjoner kan brukes for å finne ord som opptrer hyppigere enn ellers. Da gjerne sammen med en referanse. Referansemodulen er ikke helt på plass, men det finnes en inne i dhlab. Her har vi en del kode som ikke er på plass ennå. Men den kommmer!

Kollokasjoner er gitt fra korpuset omtrent som for konkordanser

``` 
dh.Collocations(
    corpus=None,
    words=None,
    before=10,
    after=10,
    reference=None,  en dataramme med fordeling av ord
)

```
I eksemplet under skal vi kun se på etterpåfølgende ord, uten referanse, og så med referanse


````python
coll = dh.Collocations(sandel, "Alberte", after = 10, before = 0)
````

````python
coll.show()
````

````python
from dhlab.api.dhlab_api import get_reference
````

````python
reference = get_reference(corpus = 'digibok', from_year = 1900, to_year = 1940)
````

````python
coll = dh.Collocations(sandel, "Alberte", after = 10, before = 0, reference = reference)
````

Med referanse i bunn, har kollokasjonen også en relevansekolonne.

````python
coll.show(sortby='relevance').head(20)
````

# Ngram

Det her er en modul for å hente ut ngrammer fra materiale, med metadata eller bare mot nb ngram.

````python
import dhlab.ngram as ng
````

På samme måte som over bygges det data, som så vises frem, eller lagres.

````python
pron = ng.Ngram(['han', 'hun'])
````

````python
pron.plot()
````

ngrammene har noen attributter, blant dem selve dataene, som også kan lagres til excel og arbeides med videre der.

````python
pron.ngram
````

````python
pron.ngram.to_excel("pronomen.xlsx")
````

For å lete spesifikt i bøker og underklasser av bøker bruk `Ngram_book`. Den gir absolutte frekvenser tilbake. Fungerer derfor best til sammenligning.

```
ng.Ngram_book(
    words=None,
    title=None,
    publisher=None,
    city=None,
    lang='nob',
    from_year=None,
    to_year=None,
    ddk=None,
    subject=None,
)
```

````python
vin_mat = ng.Ngram_book(['hvitvin', 'rødvin'], from_year = 1960, to_year = 2015)
````

````python
vin_mat.plot()
````

````python
vin_mat_avis = ng.Ngram_news(['hvitvin', 'rødvin'], from_year = 1960, to_year = 2015)
````

````python
vin_mat_avis.plot()
````
