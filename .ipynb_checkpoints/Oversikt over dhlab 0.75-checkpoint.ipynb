# Modulen `nb` 

Her ligger det meste av kommandoer for tekstbehandling som konkordans og kollokasjon, og statistikk fra bøker. 

Det ligger en del i moduler som er utviklet i etterkant også, som vi tar etter hvert. 

Det vi først og fremst skal gjøre er å hente ut informasjon om ord og tekster fra et korpus.  Korpuset vil typisk være et sett referanser til dokumenter i form av en liste av URN-er (eller reduserte URN-er). URN-ene kan være implisitt eller eksplisitt. 

Søk etter URNer er fasettert slik at forfattere markeres med 'author', emneord med 'subject', ett for hver type fasett eller parameter. Mekanismen og indekseringen for de nye metadata er noe forskjellig, men kommandoene som møter brukeren vil være noenlunde lik.

## Konkordans, kollokasjon og nøkkelord

```python
# For noen av parametrene til book_corpus bør det brukes jokertegn

korpus = nb.book_corpus(ddk = '641.2%', limit = 100)
```

```python
# det er flere kommandoer for korpus, den her gir en dataramme med grunnleggende informasjon

korpus.head(20)
```

For å gjøre søk med URN, legges de til fra datarammen sammen med et søkeord.

```python
nb.urn_concordance(word='maltpreg', urns=nb.pure_urn(korpus), before = 10, after = 10, size=2)
```

Kollokasjoner har tilsvarende kommando, som gir en liste av ord som kan bearbeides. 

```python
coll = nb.urn_coll(word='syrlig', urns =nb.pure_urn(korpus))
coll
```

```python
# normaliserer, men her er det flere muligheter for normalisering - alt avhengig av hva som skal gjøres.

nb.normalize_corpus_dataframe(coll)
```

Bearbeiding med kollokasjoner kan benytte frekvenslister for hele boksamlingen som referanse. Husk at kollokasjon kan sees på som en tolkning av formelen

$$\frac{p(w|C)}{p(w)}$$

Der $p(w)$ er sannsynligheten (eller proporsjonen) av ordet $w$ i totalkorpuset, og $p(w|C)$ er den betingede sannsylighet (proporsjon) for $w$ i ordlisten $C$, som kan komme fra en kollokasjon eller et annet korpus. Om verdien er 1  har $w$ identisk distribusjon i helheten som i korpuset.


```python
# kommandoen nb.totals() bygges inn i nb.frame() som er en forkortelse for pandas datarammekonstruktør

tot = nb.frame(nb.totals(200000), 'tot')
nb.normalize_corpus_dataframe(tot)
```

De mest relevant for syrlig i korpuset er. Merk at normaliseringen vi har vil jekke opp tallene noe, rett og slett fordi det er litt færre. Men sorteringen av tallene vil være det samme.

```python
nb.frame_sort(nb.frame(koll[0]/tot.tot)).head(20)
```

##### Kobling til formel

Uttrykket $\frac{\textrm{koll}[0]}{\textrm{tot.tot}}$ tilsvarer beregningen i $p(w) \sim \textrm{tot.tot}$ og $p(w|C) \sim \textrm{koll}[0]$. 

Samme metode for å analysere hele korpuset. Kommando/funksjon for å hente ut summen av alle ord i korpuset er `nb.aggregate_urns`

```python
korpus_total = nb.frame(nb.aggregate_urns(nb.pure_urn(korpus)))
nb.normalize_corpus_dataframe(korpus_total)
```

```python
nb.frame_sort(korpus_total).head(10)
```

```python
# Passer bra for 641.2 (alkoholholdige drikker)

korpus_keys = nb.frame_sort(nb.frame(korpus_total[0]/tot.tot))
korpus_keys.head(20)
```

## Frekvenslister, nøkkelord og grafkonstruksjoner

Frekvenslister kan hentes ut for alle bøker. URN er redusert URN.

```python
# For å få hele listen, sett top = 0 og cutoff = 0

nb.frame(nb.get_freq(nb.pure_urn(korpus)[10], top = 10, cutoff = 10))
```

Med `get_freq` kan det bygges en DTM rimelig kjapt. En åpenbar ting her, er at koden bør pakkes inn.

```python
import pandas as pd
```

```python
dtm = pd.concat([nb.frame(nb.get_freq(u, top = 0, cutoff = 0), str(u)) for u in nb.pure_urn(korpus)[:10]], axis = 1)
```

```python
dtm
```

```python
dtm_key = pd.concat([(dtm[x]/tot.tot).sort_values(ascending=False).head(150) for x in dtm], axis = 1)
```

```python
dtm_key.columns = dtm.columns
```

En måte å behandle et korpus med bruk av modulen `gnl`. Her skal vi se hvordan det kan gjøres på klientsiden, etterpå hvordan på serversiden. 

```python
dtm_key.sort_values(by='2012120738058', ascending = False)
```

```python
graph = dtm_key.fillna(0).dot(dtm_key.fillna(0).transpose())
```

```python
graph.head(20)
```

Henter ut en graph.

```python
import networkx as nx
```

```python
G = nx.from_pandas_adjacency(graph)
```

Trimmer grafen ned litt, men vi får noe ut.

```python
Gf = nx.from_edgelist([(u,v, e) for u,v,e in G.edges(data=True) if e['weight'] > 10000000000000000])
```

```python
gnl.show_graph(Gf)
```

Clustringen blir forholdsvis grov, noe som er forventet når det gjøres på boknivå, men noe får vi ut. I og med at korpuset har en viss vilkårlig karakter vil noen clustre kanskje påvirkes av enkelte bøker.

Louvain-metoden gjør det mulig å lage subclustre om nødvendig. Nedenfor skal vi ta de viktigste nøkkelordene for korpuset og ta kryssprodukt basert på samforekomst i avsnitt.

```python
gnl.show_communities(Gf)
```

En annen måte å generere grafer på er å benytte serversiden. For øyeblikket er det en kommando for det `make_graph`, som tar et sett ord, og lager en graf for en bok. Grafene for flere bøker kan kobles sammen.

De viktigste ordene ligger i toppen variabelen `korpus_keys` definert over.

```python
korpus_keys.head(10)
```

Om vi velger de 200 på topp i den listen skulle vi være rimelig trygg.

```python
korpus_keys.head(200).iloc[199]
```

Lager en graf for ti bøker. Grafene konstrueres på serversiden.

```python
Gs = [nb.make_network_graph(u, list(korpus_keys.head(200).index)) for u in nb.pure_urn(korpus)[:10]]
```

```python
gnl.show_graph(Gs[0])
```

```python
gnl.show_communities(Gs[0])
```

```python
gnl.show_communities(Gs[1])
```

```python
gnl.show_communities(Gs[2])
```

Her kan vi prøve å bruke networkx sin egen merge-kommando, men den klarer ikke å legge sammen vekter. 

```python
F = nx.compose_all(Gs)
```

```python
gnl.show_graph(F, spread = 1.8)
```

Clustrene her ser rimelige ut med tanke på semantiske felt. Altså en litt generell om viner og, og en om druetyper, med iblandet litt adjektiv, en med områder, så litt om vinmarker, og diverse. Louvain kan være litt random, så både antall og innholdi clustrene kan variere. 

Grafen er er det som kommer ut ved å kombinere graf-konstruksjon for de viktigste ordene fra korpuset. Alt gjort per URN på serverside med kommandoen `make_network_graph`, basert på samforekomst av ord i avsnitt .

```python
gnl.show_communities(F)
```
