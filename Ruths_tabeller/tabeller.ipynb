```python
table = dict()
for row in  rows.replace(', ', '').split('%'):
    if row != '':
        header_data = re.split("#", row)
        header = header_data [0].strip()
        rowcells = header_data [1].strip()
        cells = list()
        for cell in re.split('\n\n', rowcells): 
            cellcontent = cell.split(':')
            cells.append((cellcontent[0], int(cellcontent[1])))
        table[header] = dict(cells)
        
```
