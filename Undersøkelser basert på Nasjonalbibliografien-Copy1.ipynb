# Oversatt engelsk litteratur

```python
eng = nb.book_corpus(period=(1950, 2010), trans='eng', limit=50000)
```

```python
eng.sort_values(by='year').groupby('year').count()['urn'].plot(lw=5, figsize=(10,5))
```

```python
alt = nb.book_corpus(period=(1950, 2010), ddk = '8%', limit = 500000)
```

```python
altm = nb.metadata(list(alt.urn))
```

```python
nb.frame(altm)
```

```python
språk = nb.frame(altm)[[0, 2, 8, 9]].groupby([2,9]).count()
```

```python
språk.plot(figsize=(15,6))
```
