## Test ut med kommando `relaterte_ord`

Det er en egenvektorbasert måte å hente ut tilsvarende ord fra grafen gitt under.

```python
def cluster_for(w, cutoff = 12, lang = 'nob'):
    G = nb.make_graph(w, cutoff = cutoff, lang=lang)
    c = gnl.community_dict(G)
    res = []
    for x in c:
        if w in c[x]:
            res = c[x]
    return res
```

```python
def cliques_for(w, cutoff = 12, lang = 'nob'):
    G = nb.make_graph(w, cutoff = cutoff, lang=lang)
    c = gnl.make_cliques_from_graph(G.to_undirected())
    res = []
    for x in c:
        if w in c[x]:
            res = c[x]
    return res
```
