```python
konks = nb.get_urnkonk('legger', {'urns':nb.pure_urn(mykle), 'before':12, 'after':12}, html=False)
konks
```

```python
rows = konks.before + ' ' + konks.word + ' ' + konks.after
```

```python
len(rows)
```

```python
rows[20]
```

```python
cols = "token.text, token.lemma_, token.pos_, token.tag_, token.dep_, token.shape_, token.is_alpha, token.is_stop".split(', ')
```

```python
cols
```

```python
parses = pd.DataFrame(columns = cols)
```

```python
for conc in rows:
    doc = nlp(conc)
    row = []
    for token in doc:
        row.append((token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop))
    parses = parses.append(pd.DataFrame(row, columns = cols))
```

```python
def parse(string):
    doc = nlp(string)
    rows = [(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop) for token in doc]
    return pd.DataFrame(rows, columns = cols), doc
```

```python
p, d = parse("leggene hans var formidable")
p
```

```python
displacy.render(d, style="dep" )
```

```python
sandel = nb.book_corpus(author='Sandel%Cora%', limit=5)
sandel
```

```python
# Her legger vi setningene i en variabel som vi kan arbeide videre med

sandel_setn = nb.sentences(nb.pure_urn(sandel), num=20)
sandel_setn
```

```python
# Skriv setningene ut til en .txt-fil

with open('setninger.txt', 'w', encoding='utf-8') as f:
    for setning in sandel_setn:
        f.write(setning)
        f.write('\n')
```
