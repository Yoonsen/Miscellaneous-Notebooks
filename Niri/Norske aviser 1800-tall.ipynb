## ocr

Et stort problem med 1800-tallet er OCR-feil. Men det er likevel mulig å orientere seg en del, og også hente ut statistikk i form av kollokasjoner. Det er to redskaper som er fine å bruke, wildcardsearch og konkordans. Konkordansene gir en lenke til bokhylla. Klikk på lenken og du kommer til første treff på ordet det søkes etter.

Husk at søkene er følsom for kapitalisering, så 'opinion' vil ha andre treff enn 'Opinion'. Substantiv på 1800-tallet vil vanligvis skrives med stor bokstav.

Resultatet av en konkordans vil gjerne være forskjellig fra søk til søk om det er mange treff. Maks for 'before' og 'after' er 12. For å søke i alle aviser er det bare å sette title='%', eller utelate parametret helt.

```python
nb.concordance(word="Opinion", corpus='avis', title="aftenposten", yearfrom=1820, yearto=1890, before = 8, after = 8, size=10)
```

```python
# her et søk på allmennvilje, men se nedenfor for skrivemåte

nb.concordance(word="Almenvillie", corpus='avis', title="%", yearfrom=1820, yearto=1890, size=10)
```

# Wildcard

For å finne spesielle OCR-feil, eller andre stavelsesvarianter, kan du bruke wildcardsøket. Ordene du finner her kan settes inn i et konkordansesøk.

Selve søket er satt opp sånn at du velger en ramme for et ord, og setter inn * der bokstavene kan være noe annet, for feil gjelder det typisk æ, ø, å, men også e og c kan byttes om, k og t kan forveksles, og stor I og J byttes gjerne om. I tillegg kommer varianter som Q og K som endret seg i perioden. Her er det bare å prøve seg frem. Selv om dataene er fra bøker, vil de også ha en relevans for avissøk.

```python
nb.sorted_wildcardsearch({
    'word': 'Al*m*v*e', # formen det søkes etter her en sjekk på allmennvilje
    'freq_lim': 50,     # minste frekvens ordet skal ha
    'limit': 50,        # antall treff som skal rapporteres, eller lengden på listen
    'factor': 4         # hvor langt ordet skal være sammenlignet med inputformen 4 bokstaver i tillegg.
})
```
