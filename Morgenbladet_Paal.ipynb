# N-gram fra fulltekster

Her noen bigrammer fra skjÃ¸nnlitteratur, eller det som er kodet som 839 i dewey-systemet

```python
period = (1850, 2020)
```

```python
komma_men = nb.bigram(',','men', ddk="839%", period= period)
men = nb.unigram('men', ddk="839%", period = period)
fordi = nb.unigram('fordi', ddk="839%", period = period)
komma_fordi = nb.bigram(',', 'fordi', ddk="839%", period = period)
```

```python
komma_men_all = nb.bigram(',','men',  period=period)
men_all = nb.unigram('men',  period = period)
fordi_all = nb.unigram('fordi', period = period)
komma_fordi_all = nb.bigram(',', 'fordi', period = period)
```

```python
punkt_men = nb.bigram('.','Men', ddk="839%", period=period)
Men = nb.unigram('Men',  ddk="839%", period=period)
punkt_fordi = nb.bigram('.','Fordi', ddk="839%", period=period)
Fordi = nb.unigram('Fordi',  ddk="839%", period=period)
```

```python
punkt_men_all = nb.bigram('.','Men',  period=period)
Men_all = nb.unigram('Men',  period=period)
punkt_fordi_all = nb.bigram('.','Fordi',  period=period)
Fordi_all = nb.unigram('Fordi',  period=period)
```

## Forholdstall

Forholdstall for bruken av _men_ og _fordi_ med og uten komma foran.

## men

Forholdet mÃ¥les som antallet _men_ med komma dividert pÃ¥ det totale antallet _men_. Kurven peker oppover, som tyder pÃ¥ at kommabruken tiltar for _men_. Litt pussig at den skulle havne over 1, som i prinsippet skal vÃ¦re umulig (hmmm).

```python
(komma_men[0]/men[0]).rolling(window=5).mean().plot(lw=3, figsize=(10,5), title=", men/men", legend=False);
(b, top) = plt.ylim()
plt.ylim((0, top+0.3));
```

## fordi

For _fordi_ er det motsatt. Kurven peker nedover, sÃ¥ trender er Ã¥ bruke _fordi_ uten komma foran.

```python
(komma_fordi[0]/fordi[0]).rolling(window=5).mean().plot(lw=3, figsize=(10,5), title=", fordi/fordi", legend=False);
(b, top) = plt.ylim()
plt.ylim((0, top*1.3));
```

## men i alle bÃ¸ker

Omtrent det samme som for 839 (skjÃ¸nnlitteratur)

```python
(komma_men_all[0]/men_all[0]).rolling(window=5).mean().plot(lw=3, figsize=(10,5), title=", men/men", legend=False);
(b, top) = plt.ylim()
plt.ylim((0, top*1.3));
```

## fordi i alle bÃ¸ker

Tilsvarende som for skjÃ¸nnlitteraturen.

```python
(komma_fordi_all[0]/fordi_all[0]).rolling(window=5).mean().plot(lw=3, figsize=(10,5), title=", fordi/fordi", legend=False);
(b, top) = plt.ylim()
plt.ylim((0, top+0.3));
```

# Fordi og Men - stor bokstav i skjÃ¸nnlitteraturen
