```python
def get_url(urn, page=1, part=200):
    # urn as digit
    urn = "URN:NBN:no-nb_digibok_" + str(urn) + '_{0:04d}'.format(page)
    print(urn)
    url = "https://www.nb.no/services/image/resolver/{urn}/full/0,{part}/0/native.jpg".format(urn = urn, part=part)
    return url
```
