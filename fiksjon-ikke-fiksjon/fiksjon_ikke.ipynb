```python
import dhlab.nbtext as nb
```

```python
import pandas as pd
```

```python
data = pd.read_excel('Oversikt norske romaner 1850-1899.xlsx')
data
```

```python
urner = nb.pure_urn(' '.join(data.URN))
```

```python
len(urner)
```

```python
agg = nb.aggregate_urns(urner)
```

```python
df = nb.frame_sort(nb.frame(agg, 'frekv'))
```

```python
df[df.frekv > 5]
```

```python
tags = {w:nb.word_form(w) for w in list(df.index)}
tags
```

```python
nb.frame(tags).to_csv('tagget_korpus.csv')
```
