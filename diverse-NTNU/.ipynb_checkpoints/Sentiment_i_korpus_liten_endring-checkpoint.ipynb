```python
import dhlab.nbtext as nb
import pandas as pd
```

```python
korpus45 = nb.restore_metadata_from_excel('korpus4554.xlsx')
korpus45
```

```python
corpus = nb.pure_urn(korpus45)
corpus
```

```python
nb.urn_concordance(word = ['tysk'], urns =corpus, before = 10, after = 10, size = 20)
```

```python
nb.frame(nb.metadata(corpus))
```

```python
tysk_ = "tysk tyske tyskere tyskerne".split()
russisk_ = "russisk russiske russere".split()
soldat_ = "soldat soldater soldatene".split()
glad_ = "fornøyd lykkelig morsom hyggelig snill vennlig hjelpsom god blid".split()
```

### Endret litt



```python
# Legger til en try-except, for de tilfellene der nb.plot... ikke fungerer
# la til alle ordistene for den her kjøringen - men ser at dere har en annen løsning, bare sett det tilbake igjen.
# legg inn try - except på samme måte nedenfor. Litt klipp og lim!

pl1 = []
for x in corpus:
    try:
        pl1.append(nb.plot_book_wordbags(x, [tysk_, russisk_, soldat_, glad_], window=50000, pr=1000))
    
    except ValueError:
        print("Problemer med:", x)
```

```python
# plot hver enkelt figur eller alle (om alle plottes vil det lage en advarsel - bare se bort ifra den)
# husk at alle ordene kan stå i wordbag...

pl1[1].plot()
```

```python
1
pl2 = []
for x in corpus:
    pl2.append(nb.plot_book_wordbags(x, [glad_], window=50000, pr=1000))
```

```python
pl3 = []
for x in corpus:
    pl3.append(nb.plot_book_wordbags(x, [soldat_], window=50000, pr=1000))
```

### Forklaring: 
Grunnen til at vi ikke bare kjører pl, men pl1, så pl2, så pl3 for hvert ord, er fordi grafene ble slått sammen/det kom flere linjer for hvert ord i grafen (eller noe i den duren). Derfor må vi egentlig lage tre grafer, som vi til slutt slår sammen med .join

```python
plott1 = pd.concat(pl1, ignore_index=True)
```

```python
pl1
```

```python
plott2 = pd.concat(pl2, ignore_index=True)
```

```python
plott3 = pd.concat(pl3, ignore_index=True)
```

```python
plott= plott1.join(plott2).join(plott3)
```

```python
plott
```

```python
plott.plot(figsize=(15,5))
```
