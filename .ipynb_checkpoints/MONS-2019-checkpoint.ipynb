```python
def dobbel(word, det='den'):
    dobbel = nb.nb_ngram('{w}, {det} {w}'.format(w=word, det=det))
    nb.frame(dobbel['{det} {w}'.format(w=word, det=det)]/dobbel[word]).plot()
```

```python
dobbel('bilen')
```
