# nb.difference()

Cascaded division. The command `difference(A, B, C, D)` computes 

$$\textrm{difference}(A,B,C,D) = \frac{\frac{A}{B}}{\frac{C}{D}}$$

So it compares the ratio of the first two arguments with the ratio of the second two. For a certain selection of arguments, this leads to a simple interpretations. If *difference* is less than 1, it means the numerator ratio is smaller than the denominator, which can be used to compare contexts, and  see how contexts make a difference. For example is *w* and *v* are two words, and *C* is a context, *difference* can be used to see if the context is associated with *w* or *v*:

If $\textrm{difference}(C(w),C(v),w,v) > 1$ then $C$ is associated with $w$, otherwise with $v$. In this case, we can say that the context $C$ prefers $w$ for $v$. 

```python
# measure difference between men and women
nb.difference('han gråter', 'hun gråter', 'han', 'hun')
```

```python
# measure difference between men and women
nb.difference('han føler', 'hun føler', 'han', 'hun')
```

```python
# measure difference between men and women
nb.difference('han påstår', 'hun påstår', 'han', 'hun')
```

```python
# measure difference between men and women
nb.difference('han hevder', 'hun hevder', 'han', 'hun')
```

At this point it may be helpful to just add in the pronouns in a template and study different contexts

```python
diff = lambda x: nb.difference("han " + x, "hun " + x, "han", "hun")
diffseries = lambda s: pd.Series(list(map(diff, s)), index = s)
```

The examples above can be formulated by specifying only the contexts

```python
he_she = diffseries("gråter påstår føler hevder sover arbeider snorker kjører".split())
```

```python
he_she.plot(kind='bar', figsize = (10,5), alpha = 0.6, rot = 0, fontsize=14 )
plt.axhline(y=1, linewidth=3, color='orange', alpha=0.6)
```

In the case above, it is clear that $\textrm{'hun gråter'}$ (she sobs) has a much higher frequency than expected, by looking at the ratio $\frac{\textrm{'han'}}{\textrm{'hun'}}$.

```python
nb.difference('rike menn', 'rike kvinner', 'menn', 'kvinner')
```

## Corpus

How to get data out of a corpus

Searching using live spec.

```python
nb.get_konk('jern', {'size':5, 'before':8, 'after': 8, 'yearfrom':1980, 'yearto':2070})
```

Specify more metadata:

```python
nb.get_konk('fiskeskinn', {'size':5, 'yearfrom':1950, 'ddk':"641%", 'yearto':2000})
```

```python
nb.get_konk('fiskeskinn', {'size':5, 'yearfrom':1950, 'author': 'juelsen%', 'ddk':"641%", 'yearto':2000})
```

Build a specific corpus:

```python
hamsun = nb.get_urn({'author':"hams%kn%", 'year':1920, 'next':100, 'limit':100})
```

We may have a look at the data

```python
hamsun
```

Check for duplicates by collecting all works with the same title in a bag (dictionary):

```python
duplicates = dict()
for meta in hamsun:
    if meta[2] in duplicates:
        duplicates[meta[2]].append(meta)
    else:
        duplicates[meta[2]] = [meta]
```

```python
duplicates
```

Sort duplicates by year, descending

```python
for item in duplicates:
    duplicates[item].sort(key = lambda x: int(x[3]), reverse=True)
```

Display in a nice way:

```python
for title in duplicates:
    printmd("### " + title )
    for item in duplicates[title]:
        printmd("* " + ' '.join([str(x) for x in item]))
```

Sorted by variants lets us construct a new corpus with unique works

```python
hamsun_unique = [duplicates[item][0] for item in duplicates]
```

This selects the newest editions. Another sorting can select the first editions.

```python
hamsun_unique
```

A concordance from a defined corpus

```python
nb.get_urnkonk('elskede',{'urns':hamsun_unique})
```

# Keywords within a corpus

Here we go about finding keywords in a corpus. First, build a corpus as a collection of URNs.

Define a corpus using metadata, like Deweys decimal system. [Webdewey](http://deweysearchno.pansoft.de/webdeweysearch/index.html) is good for that.

```python
korpus = nb.get_urn({'ddk':"355%", 'limit':200, 'year':1920, 'next':130})
```

```python
len(korpus)
```

Inspect the corpus using a dataframe.

```python
korpus_df = nb.frame(nb.metadata(korpus))
korpus_df.columns = ['urn', 'author', 'year', 'title', 'publisher', 'lang', 'trans']
korpus_df
```

Fetching corpus as a document-term matrix

```python
corpus_text = nb.get_corpus_text([x[0] for x in korpus])
```

Study the DTM (Document Term Matrix) using dataframes displayed with heatmaps

```python
nb.heatmap(
    corpus_text.sort_values(by = corpus_text.columns[0], ascending=False)[corpus_text.columns[:5]][:10]
)
```

The book with URN=2014082908006 looks a bit suspiscious. Check it out using the korpus dataframe, or simply do a new lookup with metadata function

```python
# data frame lookup

korpus_df[korpus_df['urn'] == 2014082908006]
```

```python
# using metadata command from nbtext

nb.metadata(2014082908006)
```

# Corpus keywords

Divide the corpus on a total, another reference corpus, or some other wordset. 

One possibibility is the grand total of all words

```python
totalen = nb.frame(nb.totals(25000))
totalen.head()
```

Normalize to make computations easier to follow.

```python
nb.normalize_corpus_dataframe(totalen)
```

Now the numbers are expressed as percentages

```python
totalen.head()
```

We aggregate the corpus and normalize. Actually, we normalize, and aggregate as means

```python
nb.normalize_corpus_dataframe(corpus_text)
```

```python
corpus_agg = corpus_text.mean(axis=1)
```

```python
corpus_agg = nb.frame(corpus_agg).sort_values(by=0, ascending=False)
```

```python
corpus_agg.head()
```

```python
corpus_keywords = nb.frame(corpus_agg[0]**1/totalen[0]).sort_values(by=0, ascending=False )
```

```python
corpus_keywords.head(20)
```

```python
nb.cloud(corpus_keywords.head(150)/corpus_keywords.head(150).sum())
```

# Collocations

A collocation is like a concordance, but we don't see the actual sequence. We just collect the words in the context and tally up.

```python
coll = nb.urn_coll('kaffe', urns = korpus, before = 5, after = 5)
```

```python
coll.sort_values(by=0, ascending=False).head(20)
```

Formally a collocation is just a bag of words, just like the corpus itself after aggregation. So the method is the same.

```python
diff = (coll**1.01/totalen).sort_values(by=0, ascending = False)
```

```python
diff.head(20)
```

```python
nb.cloud(diff[:100]/diff[:100].sum())
```

```python
coll2 = nb.urn_coll('offiser', urns = korpus, before = 5, after = 5)
```

```python
coll2.sort_values(by=0, ascending=False).head(20)
```

Formally a collocation is just a bag of words, just like the corpus itself after aggregation. So the method is the same.

```python
diff2 = (coll2**1.01/totalen).sort_values(by=0, ascending = False)
```

```python
diff2.head(20)
```

```python
nb.cloud(diff2[:100]/diff2[:100].sum())
```
