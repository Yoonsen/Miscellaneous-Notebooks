```python
import requests
import csv
from bs4 import BeautifulSoup
import os


def text_(body):
    """Extract text from html page from Goodreads"""
    soup = BeautifulSoup(body, 'html.parser')
    texts = soup.find_all('span', attrs={'id':True})
    #visible_texts = filter(tag_visible, texts)  
    return texts

def get_files(mappe):
    """get files, but only at the first level (note use of next())"""
    files =[
        os.path.join(mappe, f) for f in  [
            f for f in next(os.walk(mappe))
        ][2]
    ]
    return files
```

```python
def extract_html_from_files(files):
    htmls = []
    for fname in files:
        with open(fname, encoding="utf-8") as f:
            htmls.append(f.read())
    return htmls
```

```python
def get_user_comments_goodread(htmls):
    """Get reviews from Goodreads html pages, and return a list of review texts"""
    for html in htmls:
        soup = BeautifulSoup(html, 'html.parser')
        divs = soup.find_all('div', attrs={'id':True})
        reviews = dict()
        for div in divs:
            if div['id'].startswith('review'):
                spans = div.find_all('span', {'id':True})
                for span in spans:

                    # Check if there is a review is more than meets the eyes - the one got by clicking the 'more'-button
                    if span.has_attr('style') and span['id'].startswith('freeText') and span['style'] == 'display:none':
                        #print("fullrev")
                        reviews[span['id']] = span.text

                    # if not the review is contained in the displayed text
                    elif span['id'].startswith('freeTextContainer'):
                        #print('delrev')
                        reviews[span['id']] = span.text
    return reviews

def get_bokomtaler_bokelskere(htmls):
    reviews = []
    reviews = []
    for html in htmls:
        soup = BeautifulSoup(html, 'html.parser')
        divs = soup.find_all('div', attrs={'class':True})
       
        for div in divs:
            divtext = ""
            if div['class'] == ['notattekst']:
                for p in div.find_all('p'):
                    divtext += p.text + "\n"
                reviews.append(divtext)
    return reviews
```
